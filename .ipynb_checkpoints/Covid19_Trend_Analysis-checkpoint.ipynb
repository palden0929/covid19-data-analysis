### 2. Cleaning the Data

```python
# Dealing with the continents listed as countries
invalid_countries = [
    'World', 'Asia', 'Europe', 'Africa', 'North America', 
    'South America', 'Oceania', 'European Union', 'International'
]

covid_subset = covid_subset[~covid_subset['country'].isin(invalid_countries)]
covid_subset = covid_subset[covid_subset['continent'].notna()]  # removes rows with no continent
```

```python
# Removing tiny countries as well as they skew the data and are not important for our data analysis
covid_subset = covid_subset[covid_subset['population'] > 100000]  # remove microstates
```

```python
# Checking the missing values left
covid_subset.isna().sum()
```

```python
log = ['total_cases', 'total_deaths', 'total_vaccinations', 'people_vaccinated', 'population']
for col in log:
    covid_subset[f'log_{col}'] = np.log1p(covid_subset[col])
```
